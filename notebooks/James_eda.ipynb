How does income correlate with specific components of the cost of living?

-Compared to the housing cost percentage and the tax rate across regions, the average monthly income around the globe seems to have stagnated. 

-In relation to the cost of living amongst other regions Africa and South America seems to have the lowest tax average. My recommendation on for a consultant looking to relocate overseas points to Africa for the simple fact that when planned correctly, this decision cost loud to a saving percentage which is evident in the conclusions of the dataset. 

```python
components=["Housing_Cost_Percentage", "Healthcare_Cost_Percentage", "Education_Cost_Percentage", "Tax_Rate", "Savings_Percentage"] 

reg_costOFLife = clean_df.groupby('Region')[components].mean().reset_index()
reg_avecostOFLife = reg_costOFLife.melt(id_vars='Region', var_name='components', value_name='OnAverage')

plt.figure(figsize= (14,8))
sns.barplot(data=reg_avecostOFLife , x='Region', y='OnAverage', hue='components')
plt.title("Cost of life across regions")
plt.xlabel("Region")
plt.ylabel("CostOnAverage")
plt.xticks(rotation=45)
plt.legend(title='Cost of life')
plt.title('Cost of Liiving Across Regions')
plt.tight_layout
plt.show()
```

```python
plt.figure(figsize=(14,8))
sns.boxplot(data=clean_df, x="Region", y='Average_Monthly_Income', hue='Region', palette='coolwarm')
plt.title("Average Monthly Income by Region")
plt.tight_layout()
plt.show()
```
