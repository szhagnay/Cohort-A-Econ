```python
#Importlibraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
#load data base 
COL = pd.read_csv("../Cohort-A-Econ-main/data/Cost_of_Living_and_Income_Extended.csv")
USM = pd.read_csv("../Cohort-A-Econ-main/data/US_macro_micro_index.csv")
```

```python
#clean data
COL = COL[COL["Country"] == "United States"]


COL.head()
```

```python
USM.drop(columns=["Unnamed: 0"])
```

```python
USM.head()
```

```python
#merge databases on year
db = pd.merge(COL, USM, on="Year")
db.head()
```

```python
#With a focus on the United States, how does the distribution of income compare to the distribution of cost of living? 
#Visualize
db.plot(
    x="Total_Personal_Income",
    y="Cost_of_Living",
    kind = "bar",
    title = "Americans total income vs cost of living",
    xlabel = "Total annual income",
    ylabel = "Cost of Living"
)

```

```python

test = np.corrcoef(db["Total_Personal_Income"], db["Cost_of_Living"])[0,1]
print(test)
```

```python
plt.figure(figsize=(13, 5))
db = db.select_dtypes(include='number') 
corr = db.corr()
sns.heatmap(corr, annot=True, cmap='viridis')
```

```python
plt.figure(figsize=(30,14))
sns.lineplot(data= db,x="Education_Cost_Percentage",y= "Savings_Percentage", marker='s', markersize=5,linewidth=3)

plt.tight_layout()
plt.show()
```
